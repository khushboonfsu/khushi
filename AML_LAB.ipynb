```python
# STEP 2: Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 3: Load Dataset (adjust delimiter and header if needed)
filename = 'WISDM_ar_v1.1_raw.txt'  # Replace with your actual file name
df = pd.read_csv(filename, delim_whitespace=True, header=None, comment=';')
df = df[0].str.split(',', expand=True)

# Explicitly name the columns
num_cols = df.shape[1]  # Get the number of columns
df.columns = [str(i) for i in range(num_cols)] # give columns names from 0 to num_cols -1

# Preview data
print("Dataset shape:", df.shape)
df.head()
```

```python
# STEP 4: Preprocessing

# If last column is activity label, separate features (X) and labels (y)
# Assuming the last column is the label after the split
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Convert labels to numeric if they are strings (e.g., "Walking")
if y.dtype == object:
    le = LabelEncoder()
    y = le.fit_transform(y)
    class_names = le.classes_
else:
    class_names = ['Walking', 'Jogging', 'Upstairs', 'Downstairs', 'Sitting', 'Standing']

# Ensure X contains only numeric values
X = X.apply(pd.to_numeric, errors='coerce') # Convert columns to numeric, handling errors

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X) # Apply scaler to the numeric features

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# STEP 5: Model Training
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# STEP 6: Classification & Evaluation
y_pred = model.predict(X_test)

# Get unique class labels from predictions
unique_classes = np.unique(y_pred)

# Adjust class_names to match predicted classes
class_names_adjusted = [class_names[i] for i in unique_classes]

print("✅ Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# STEP 7: Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
